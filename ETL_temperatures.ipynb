### Import CSV file

```python
import pandas as pd
```

```python
csv_file = "nycavgtemps.csv"

avg_temps = pd.read_csv(csv_file)
avg_temps.head()
```

### Find average temperature for each season

```python
avg_temps["Winter"] = round((avg_temps["JAN"] + avg_temps["FEB"] + avg_temps["MAR"])/3,1)
avg_temps["Spring"] = round((avg_temps["APR"] + avg_temps["MAY"] + avg_temps["JUN"])/3,1)
avg_temps["Summer"] = round((avg_temps["JUL"] + avg_temps["AUG"] + avg_temps["SEP"])/3,1)
avg_temps["Fall"] = round((avg_temps["OCT"] + avg_temps["NOV"] + avg_temps["DEC"])/3,1)
avg_temps
```

### Only keep data from 1990-2015 to match emissions dataset

```python
avg_temps = avg_temps.iloc[4:26]
avg_temps
```

### Set index to year

```python
avg_temps = avg_temps.set_index("YEAR")
avg_temps = avg_temps[["ANNUAL", "Winter", "Spring", "Summer", "Fall"]]
avg_temps
```
