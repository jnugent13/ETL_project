### Import CSV file

```python
import pandas as pd
```

```python
csv_file = "nycavgtemps.csv"

avg_temps = pd.read_csv(csv_file)
avg_temps.head()
```

### Find average temperature for each season

```python
avg_temps["Winter"] = round((avg_temps["JAN"] + avg_temps["FEB"] + avg_temps["MAR"])/3,1)
avg_temps["Spring"] = round((avg_temps["APR"] + avg_temps["MAY"] + avg_temps["JUN"])/3,1)
avg_temps["Summer"] = round((avg_temps["JUL"] + avg_temps["AUG"] + avg_temps["SEP"])/3,1)
avg_temps["Fall"] = round((avg_temps["OCT"] + avg_temps["NOV"] + avg_temps["DEC"])/3,1)
avg_temps
```

### Only keep data from 1990-2015 to match emissions dataset

```python
avg_temps = avg_temps.iloc[4:26]
avg_temps
```

### Set index to year

```python
avg_temps = avg_temps.set_index("YEAR")
avg_temps = avg_temps[["ANNUAL", "Winter", "Spring", "Summer", "Fall"]]
avg_temps
```

## New York Green House Gas data

```python
# Import the data in Pandas
csv_path = "NY_greenhouse_gas.csv"
ghgas_df = pd.read_csv(csv_path)
```

```python
ghgas_df.head()
```

```python
# EPA combins Electric Generation and Net Imports of Electricity as Electricity production.
# So, sum values in these two columns for each year, and create a new column called as Electricity_production

Electricity_production=ghgas_df['Electric Generation Total'] + ghgas_df['Net Imports of Electricity']
Electricity_production
```

```python
# Add the new column to the table

ghgas_df['Electricity_production'] = Electricity_production
ghgas_df.head()
```

```python
# We don't need Electric Generation Total and Net Imports of Electricity, so delete them
# reference https://www.geeksforgeeks.org/python-delete-rows-columns-from-dataframe-using-pandas-drop/

ghgas_df2=ghgas_df.drop(["Electric Generation Total", "Net Imports of Electricity"], axis=1) 
ghgas_df2.head()
```

```python
# Change column order and change the column headers to uppercase

ghgas_df2=ghgas_df[["Year", "Residential Total", "Commercial Total", "Industrial Total", "Transportation Total", "Electricity_production", "Year Total"]]

ghgas_df2.rename(columns = {'Year':'YEAR', 'Residential Total':'Residential_Total', 
                            'Commercial Total':'Commercial_Total', 'Industrial Total':'Industrial_Total', 
                            'Transportation Total':'Transportation_Total', 'Electricity_production':'Electricity_Production',
                            'Year Total':'Year_Total'}, inplace = True) 

ghgas_df2.columns = ghgas_df2.columns.str.upper()

ghgas_df2.head()
```

```python
# making the year (column) as index

ghgas_df3 = ghgas_df2.set_index('YEAR')
ghgas_df3
```

```python
# Merge two tables on year

merged_data=ghgas_df3.merge(avg_temps, how='inner', on='YEAR')
merged_data
```
