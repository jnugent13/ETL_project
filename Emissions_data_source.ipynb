### Import csv file

```python
import pandas as pd
```

```python
csv_file = "ny_emissions_source.csv"
emissions_df = pd.read_csv(csv_file)
emissions_df.head()
```

### Filter for NY data

```python
emissions_df = emissions_df.loc[emissions_df["STATE_ABBR"] == "NY", :]
emissions_df
```

### Clean up columns

```python
emissions_df = emissions_df.drop(columns = ["STATE_FIPS", "STATE_ABBR"])
emissions_df = emissions_df.rename(columns = {"emissions90": "1990",
                                             "emissions96": "1996",
                                             "emissions97": "1997",
                                             "emissions98": "1998",
                                             "emissions99": "1999",
                                             "emissions00": "2000",
                                             "emissions01": "2001",
                                             "emissions02": "2002",
                                             "emissions03": "2003",
                                             "emissions04": "2004",
                                             "emissions05": "2005",
                                             "emissions06": "2006",
                                             "emissions07": "2007",
                                             "emissions08": "2008",
                                             "emissions09": "2009",
                                             "emissions10": "2010",
                                             "emissions11": "2011",
                                             "emissions12": "2012",
                                             "emissions13": "2013",
                                             "emissions14": "2014",
                                             "emissions15": "2015",
                                             "emissions16": "2016",
                                             "emissions17": "2017"})
emissions_df.head()
```
