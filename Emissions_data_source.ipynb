### Import csv file

```python
import pandas as pd
```

```python
csv_file = "ny_emissions_source.csv"
emissions_df = pd.read_csv(csv_file)
emissions_df.head()
```

### Filter for NY data

```python
emissions_df = emissions_df.loc[emissions_df["STATE_ABBR"] == "NY", :]
emissions_df
```

### Clean up columns

```python
emissions_df = emissions_df.drop(columns = ["STATE_FIPS", "STATE_ABBR"])
emissions_df = emissions_df.rename(columns = {"emissions90": "1990",
                                             "emissions96": "1996",
                                             "emissions97": "1997",
                                             "emissions98": "1998",
                                             "emissions99": "1999",
                                             "emissions00": "2000",
                                             "emissions01": "2001",
                                             "emissions02": "2002",
                                             "emissions03": "2003",
                                             "emissions04": "2004",
                                             "emissions05": "2005",
                                             "emissions06": "2006",
                                             "emissions07": "2007",
                                             "emissions08": "2008",
                                             "emissions09": "2009",
                                             "emissions10": "2010",
                                             "emissions11": "2011",
                                             "emissions12": "2012",
                                             "emissions13": "2013",
                                             "emissions14": "2014",
                                             "emissions15": "2015",
                                             "emissions16": "2016",
                                             "emissions17": "2017"})
emissions_df.head()
```

### Export to PostgreSQL

```python
from sqlalchemy import create_engine

# Imports the methods needed to abstract classes into tables
from sqlalchemy.ext.declarative import declarative_base

# Allow us to declare column types
from sqlalchemy import Column, Integer, String, Float, Date

import psycopg2
```

```python
import sqlalchemy  # Package for accessing SQL databases via Python
# Reference http://www.jan-langfellner.de/storing-a-pandas-dataframe-in-a-postgresql-database/
# Connect to database (Note: The package psychopg2 is required for Postgres to work with SQLAlchemy)
engine = sqlalchemy.create_engine('postgres+psycopg2://postgres:sp3y3grl+1@localhost:5432/ETL_project')
con = engine.connect()
```

```python
print(engine.table_names())
```

```python
# Reference https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_sql.html
# Reference https://stackoverflow.com/questions/41740985/how-do-i-insert-a-pandas-dataframe-to-an-existing-postgresql-table
emissions_df.to_sql('emissions_type', con=engine, if_exists='replace')
engine.execute("SELECT * FROM emissions_type").fetchall()
```
